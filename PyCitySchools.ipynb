```python
# DATA CLEANUP BEGINS HERE

# generate list of identified prefixes and suffixes for cleanup
prefixes_suffixes = ["Mr. ", "Mrs. ", "Miss ", "Ms. ", "Dr. ", " DDS", " DVM", " MD", " PhD"]
print(prefixes_suffixes)
```

```python
# Iterate through prefixes and suffixes and remove all usages from data set

for word in prefixes_suffixes:
    student_df["student_name"] = student_df["student_name"].str.replace(word,"")

student_df.head(10)

# DATA CLEANUP ENDS HERE
```

```python
# ANALYSIS BEGINS HERE

# Combine school and student dataframes into single data set
school_data_complete_df = pd.merge(student_df, school_df, on=["school_name", "school_name"])
school_data_complete_df.head(10)
```

```python
# identify total number of students
student_count = school_data_complete_df["Student ID"].count()
student_count
```

```python
# identify total number of schools
school_count = school_df["School ID"].count()
school_count
```

```python
# Alternate method for identifying schools
school_count_2 = len(school_data_complete_df["School ID"].unique())
school_count_2
```

```python
# determine total district budget
total_budget = school_df["budget"].sum()
total_budget
```

```python
# determine average score for reading
average_reading_score = school_data_complete_df["reading_score"].mean()
average_reading_score
```

```python
# determine average score for math
average_math_score = school_data_complete_df["math_score"].mean()
average_math_score
```

```python
# identify students with passing grades in math and reading.
# passing_reading = school_data_complete_df["reading_score"] >= 70
# passing_math = school_data_complete_df["math_score"] >= 70
```

```python
# Populate students passing math in new DataFrame
passing_math = school_data_complete_df[school_data_complete_df["math_score"] >= 70]
passing_math
```

```python
# Populate students passing reading in new DataFrame
passing_reading = school_data_complete_df[school_data_complete_df["reading_score"] >= 70]
passing_reading
```

```python
# identify counts of passing students
passing_math_count = passing_math["Student ID"].count()
passing_reading_count = passing_reading["Student ID"].count()
```

```python
# identify percentages of passing students
passing_math_pct = passing_math_count / float(student_count)
passing_reading_pct = passing_reading_count / float(student_count)                                     
```

```python
# get count of students passing both math and reading
passing_math_reading = school_data_complete_df[(school_data_complete_df["math_score"] >= 70) 
                                               & (school_data_complete_df["reading_score"] >= 70)]
passing_math_reading
```

```python
# identify count of students passing both math and reading
passing_math_reading_count = passing_math_reading["Student ID"].count()
passing_math_reading_count
```

```python
# identify percentage of students passing both math and reading
passing_math_reading_pct = passing_math_reading_count / float(student_count)
passing_math_reading_pct
```

```python
# Assemble data frame for district summary statistics.
# Use dictionary to store each kvp.
district_summary_df = pd.DataFrame(
        [
            {"Total Schools": school_count,
            "Total Students": student_count,
            "Total Budget": total_budget,
            "Average Reading Score": average_reading_score,
            "Average Math Score": average_math_score,
            "% Passing Reading": passing_reading_pct,
            "% Passing Math": passing_math_pct,
            "% Overall Passing": passing_math_reading_pct}
        ]
)
district_summary_df
```

```python
district_summary_df["Total Students"] = district_summary_df["Total Students"].map("{:,}".format)
district_summary_df["Total Budget"] = district_summary_df["Total Budget"].map("${:,.2f}".format)
district_summary_df["Average Reading Score"] = district_summary_df["Average Reading Score"].map("{:.1f}".format)
district_summary_df["Average Math Score"] = district_summary_df["Average Math Score"].map("{:.1f}".format)
district_summary_df["% Passing Reading"] = district_summary_df["% Passing Reading"].map("{:.0%}".format)
district_summary_df["% Passing Math"] = district_summary_df["% Passing Math"].map("{:.0%}".format)
district_summary_df["% Overall Passing"] = district_summary_df["% Overall Passing"].map("{:.0%}".format)
district_summary_df
```

```python
# reassign column orders
new_column_order = ["Total Schools", "Total Budget", "Average Math Score", "Average Reading Score",
                    "% Passing Reading", "% Passing Math", "% Overall Passing"]
                    
district_summary_df = district_summary_df[new_column_order]
district_summary_df
```
